# 1B: Library Book Management System

```python
from tkinter import *
import sqlite3 
import tkinter.ttk as ttk
import tkinter.messagebox as tkmessagebox

frame = Tk()
frame.title('Library Book Management Application')

#******************************************* METHODS *******************************************

def Database():
    global conn, cursor
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS library (booktitle TEXT, author TEXT, year INTEGER, ISBN INTEGER)")
    
def Displaybook():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM library ORDER BY booktitle ASC")
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
  
    cursor.close()
    conn.close()
    tkmessagebox.showinfo('Liabrary','All the Books in the Library are Displayed.')
    
def Searchbook():
    
    titleauthor=((author.get()).title())
    titlebookname=((booktitle.get()).title())
    titlepublishedyear=str(year.get())
    titleisbn=str(isbn.get())
    
    if ((titleisbn)==''):
        titleisbn='---'
        
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM library")
    fetch=cursor.fetchall()
    
    if ((titlebookname=='') and (titleauthor=='') and (titlepublishedyear=='') and (titleisbn=='')):
        tkmessagebox.showinfo('Library','Enter 1 or more Details of the book.')
    else:  
        count=0
        for data in fetch:
            if(titlebookname==data[0]) and (titleauthor==data[1]) and (str(titlepublishedyear)==str(data[2])) and (str(titleisbn)==str(data[3])):
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
            elif(titlebookname==data[0]):
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
            elif(titleauthor==data[1]):
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
            elif(str(titlepublishedyear)==str(data[2])):
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
            elif(str(titleisbn)==str(data[3]) and titleisbn !='---'):
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
            
        if(count==0):
            tkmessagebox.showinfo('Library','Book not Found')
        
    cursor.close()
    conn.close()
    author.set("")
    booktitle.set("")
    year.set("")
    isbn.set("")
    
    
def Addbook():
    
    titleauthor=((author.get()).title())
    titlebookname=((booktitle.get()).title())
    titlepublishedyear=(year.get())
    titleisbn=(isbn.get())
    
    if ((titleisbn)==''):
        titleisbn='---'

    if ((titlebookname=='') or (titleauthor=='') or (titlepublishedyear=='')):
        tkmessagebox.showinfo('Library','Enter All the Details.')
    else:
        Database()
        cursor.execute("SELECT * FROM library")
        fetch=cursor.fetchall()
        count=0
        
        for data in fetch:
            
            if(titlebookname==data[0]) and (titleauthor==data[1]) and (str(titlepublishedyear)==str(data[2])) and (str(titleisbn)==str(data[3])):
                count+=1
            
        if(count==0):
            cursor.execute("INSERT INTO library(booktitle,author,year,isbn) VALUES(?,?,?,?)",(titlebookname,titleauthor,titlepublishedyear,titleisbn))
            tkmessagebox.showinfo('Library','Values inserted')
        else:
            tkmessagebox.showinfo('Library','Book Already Exist.')
            
        conn.commit()
        cursor.close()
        conn.close()
        author.set("")
        booktitle.set("")
        year.set("")
        isbn.set("")
        
def Deletebook():
    
    titleauthor=((author.get()).title())
    titlebookname=((booktitle.get()).title())
    titlepublishedyear=str(year.get())
    titleisbn=str(isbn.get())
    
    if ((titleisbn)==''):
        titleisbn='---'
        
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM library")
    fetch=cursor.fetchall()
    
    count1=0
    count2=0
    
    for data in fetch:
        if ((titlebookname=='') or (titleauthor=='') or (titlepublishedyear=='')):
            count1+=1
            break
            
        elif ((titlebookname==data[0])and (titleauthor==data[1])and (str(titlepublishedyear)==str(data[2])) and (str(titleisbn)==str(data[3]))):
            tempa=("'"+str(titlebookname)+"'")
            tempb=("'"+str(titleauthor)+"'")
            tempp=("'"+str(titlepublishedyear)+"'")
            tempi=("'"+str(titleisbn)+"'")
            temp1="DELETE FROM library WHERE booktitle= "+tempa +"and author= "+tempb +"and year= "+tempp +"and isbn= "+tempi
            cursor.execute(temp1)
            conn.commit()
            count2+=1
            
    if(count1!=0 and count2==0):
        tkmessagebox.showinfo('Library','Enter All the Details.')
    elif(count1==0 and count2==0):
        tkmessagebox.showinfo('Library','No Book with Entered Details.')
    else:
        tkmessagebox.showinfo('Library','Book deleted')
        
    author.set("")
    booktitle.set("")
    year.set("")
    isbn.set("")
    conn.commit()
    cursor.close()
    conn.close()
    
def Exit():
    result=tkmessagebox.askquestion('Library','Do you want to exit?',icon='warning')
    
    if(result=='yes'):
        frame.destroy()
    else:
        tkmessagebox.showinfo('Library','Application screen')


#******************************************* FRAMES *******************************************

TitleFrame=Frame(frame,width=1350,bd=10,relief=RAISED,bg='grey85')
TitleFrame.pack(side=TOP)
lblTitle=Label(TitleFrame,width=50,font=('Lucida Sans',35),text='....Library Management Application....',fg='#000')
lblTitle.grid()

LeftFrame=Frame(frame,bd=10,width=930,height=520,relief=RIDGE,bg='red')
LeftFrame.pack(side=LEFT)

ButtonFrame=Frame(LeftFrame,bd=10,relief=RIDGE,bg='grey85')
ButtonFrame.pack(side=BOTTOM)

Form=Frame(LeftFrame,bd=20,width=620,height=520,bg='grey85')
Form.pack()

RightFrame=Frame(frame,bd=10,width=620,height=520,relief=RIDGE,bg='yellow')
RightFrame.pack(side=RIGHT)

#******************************************* VARIABLES *******************************************

author=StringVar()
booktitle=StringVar()
year=StringVar()
isbn=StringVar()

#******************************************* LABLE & ENTRY *******************************************

lbl_authorname=Label(Form,text="Booktitle:",font=('Rockwell',16,'bold'),fg='#000',bg='grey85')
lbl_authorname.grid(row=0,column=0,sticky='e')
entry_authorname=Entry(Form,textvariable=booktitle,width=35)
entry_authorname.grid(row=0,column=1)

lbl_bookname=Label(Form,text="Author:",font=('Rockwell',16,'bold'),fg='#000',bg='grey85')
lbl_bookname.grid(row=1,column=0,sticky='e')
entry_bookname=Entry(Form,textvariable=author,width=35)
entry_bookname.grid(row=1,column=1)

lbl_publishyear=Label(Form,text="Year:",font=('Rockwell',16,'bold'),fg='#000',bg='grey85')
lbl_publishyear.grid(row=2,column=0,sticky='e')
entry_publishyear=Entry(Form,textvariable=year,width=35)
entry_publishyear.grid(row=2,column=1)

lbl_ISBN=Label(Form,text="ISBN:",font=('Rockwell',16,'bold'),fg='#000',bg='grey85')
lbl_ISBN.grid(row=3,column=0,sticky='e')
entry_ISBN=Entry(Form,textvariable=isbn,width=35)
entry_ISBN.grid(row=3,column=1)

#******************************************* BUTTONS *******************************************

disp_button=Button(ButtonFrame,text="Display",width=9,font=('Rockwell',15),command=Displaybook,bg='grey85', fg='#000')
disp_button.grid(row=1,column=0)

search_button=Button(ButtonFrame,text="Search",width=9,font=('Rockwell',15),command=Searchbook,bg='grey85', fg='#000')
search_button.grid(row=1,column=1)

add_button=Button(ButtonFrame,text="Add",width=9,font=('Rockwell',15),command=Addbook,bg='grey85', fg='#000')
add_button.grid(row=1,column=2)

issue_button=Button(ButtonFrame,text="Issue",width=9,font=('Rockwell',15),state=DISABLED,bg='grey85', fg='#000')
issue_button.grid(row=1,column=3)

delete_button=Button(ButtonFrame,text="Delete",width=9,font=('Rockwell',15),command=Deletebook,bg='grey85', fg='#000')
delete_button.grid(row=1,column=4)

exit_button=Button(ButtonFrame,text="Exit",width=9,font=('Rockwell',15),command=Exit,bg='grey85', fg='#000')
exit_button.grid(row=1,column=5)

#******************************************* SCROLLBAR *******************************************

col=('Book_Name','Author_Name','Published_year','ISBN')
tree=ttk.Treeview(RightFrame,columns=col,show='headings')
scrollbary=Scrollbar(RightFrame,orient='vertical')
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill='y')
scrollbarx=Scrollbar(RightFrame, orient='horizontal')
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill='x')


tree.heading("Book_Name",text="Booktitle",anchor=W)
tree.heading("Author_Name",text="Author",anchor=W)
tree.heading("Published_year",text="Year",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)

tree.column("Author_Name",minwidth=0,width=250)
tree.column("Book_Name",minwidth=0,width=250)
tree.column("Published_year",minwidth=0,width=200)
tree.column("ISBN",minwidth=0,width=200)
tree.pack(side=TOP,fill=BOTH)

#******************************************* INITIALIZATION *******************************************

if __name__ == '__main__':
    frame.configure(bg='grey85')
    frame.mainloop()
```
